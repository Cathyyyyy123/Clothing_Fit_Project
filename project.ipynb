```python
def readGz(path):
  for l in gzip.open(path, 'rt'):
    yield eval(l)

def readCSV(path):
  f = gzip.open(path, 'rt')
  f.readline()
  for l in f:
    yield json.loads(l)
```

```python
allRatings = []
userRatings = defaultdict(list)

for record in readCSV("renttherunway_final_data.json.gz"):
    user = record.get("user_id")
    book = record.get("item_id")  # Assuming "item_id" corresponds to 'book'
    rating = record.get("rating")  # Get the "rating" field
    if rating is not None:  # Ensure it is not None
        try:
            r = int(rating)  # Convert to integer
            allRatings.append(r)
            userRatings[user].append(r)
        except ValueError:  # Handle invalid numeric values
            print(f"Skipping record with invalid rating: {record}")
    else:
        print(f"Skipping record with missing rating: {record}")
    allRatings.append(r)
    userRatings[user].append(r)
```
